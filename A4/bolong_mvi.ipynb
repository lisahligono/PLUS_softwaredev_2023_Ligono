<h2> Mangrove Vegetation Index </h2>
The function opens the Sentinel-2 image file located at the specified path (/Users/lisahligono/practice-plus/A4/mvi/bolong.tif) using the rasterio.open() function. The function reads the pixel values for three specific bands from the opened dataset: Near-infrared (B8), Green (B3), and Short-wave infrared 1 (B11). These bands are accessed using the read() method on the dataset object, with the band indices as arguments (3 for B3, 8 for B8, and 11 for B11). The MVI index is calculated using the formula (B8 - B3) / (B11 - B3), where B8, B3, and B11 represent the pixel values of the corresponding bands. The calculated MVI index is returned as the output of the function.
